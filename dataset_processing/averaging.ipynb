```python
data=np.genfromtxt("dataset/cfwd_split_4/chb_01_18_test.csv",delimiter=",")
```

```python
ans=[]
for i in range(len(x_train)):
    temp=[]
    for j in range(len(x_train[i])):
        temp.append(np.sum(x_train[i][j:10+j]))
    ans.append(temp)
```

```python
a=[i for i in range(len(ans[0]))]
plt.plot(a,ans[0],x_train[0][:len(a)])
```
